### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
numeric_cols = ['Age']

data['Age'].plot.hist(bins=20, title='Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
cgpa_mapping = {
    '0 - 1.99': 1.5,
    '2.00 - 2.49': 2.25,
    '2.50 - 2.99': 2.75,
    '3.00 - 3.49': 3.25,
    '3.50 - 4.00': 3.75,
}

data['CGPA'] = data['CGPA'].map(cgpa_mapping)

plt.hist(data['CGPA'], bins=5, edgecolor='black')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.title('Distribution of CGPA')
plt.show()
```
